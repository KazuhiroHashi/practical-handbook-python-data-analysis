## concat() 関数によるDataFrameの連結

```python
from IPython.display import display
import pandas as pd

df1 = pd.DataFrame(
    [
        [1, 10, 100],
        [2, 20, 200],
        [3, 30, 300],
    ],
    index=list("abc"),
    columns=list("ABC"),
)
df2 = pd.DataFrame(
    [
        [4, 10, 100],
        [2, 20, 200],
        [3, 30, 400],
    ],
    index=list("acd"),
    columns=list("ACD"),
)
display(df1, df2)
```

```python
pd.concat([df1, df2], ignore_index=True)
```

```python
labeled_df = pd.concat([df1, df2], keys=["data1", "data2"])
labeled_df
```

## concat() 関数によるDataFrameの結合

```python
pd.concat([df1, df2], axis=1)
```

```python
pd.concat([df1, df2], axis=1, join="inner")
```

## join() メソッドによるDataFrameの結合

```python
df3 = pd.DataFrame(
    [[1.3, 10.3], [2.3, 20.3]],
    index=list("ab"),
    columns=list("DE"),
)
display(df1, df2, df3)
```

```python
df1.join(df3)
```

```python
df1.join(
    df2,
    how="inner",
    lsuffix="_df1",
    rsuffix="_df2",
)
```

```python
df4 = pd.DataFrame([["a", 1.4, 10.4], ["b", 2.4, 20.4]], columns=["key", "F", "G"])
display(df4, df1)
```

```python
df4.join(df1, on="key")
```

## merge() メソッドによるDataFrameの結合

```python
pd.merge(df1, df2, on="A")
```

```python
pd.merge(df1, df2, on="A", how="left")
```

```python
pd.merge(df1, df2, left_on="C", right_on="D")
```

```python
pd.merge(
    df1,
    df2,
    on="A",
    suffixes=("_df1", "_df2"),
)
```

```python
pd.merge(
    df1,
    df2,
    on="A",
    how="outer",
    indicator=True,
)
```

```python
pd.merge(
    df1,
    df2,
    how="cross",
    suffixes=("_df1", "_df2")
)
```

```python
import numpy as np

pd.merge(
    pd.DataFrame({"a": [np.nan, 1, 2]}),
    pd.DataFrame({"a": [1, 3, np.nan, np.nan]})
)
```

### 重複したキーの検証

```python
df5 = pd.DataFrame(
    [
        [1, 10, 100],
        [2, 20, 200],
        [2, 20, 200],
    ],
    index=list("abc"),
    columns=list("ABC"),
)
df5
```

```python
pd.merge(df1, df5, on="A", validate="one_to_one")
```

```python
pd.merge(
    df1,
    df5,
    on="A",
    validate="one_to_many",
)
```
