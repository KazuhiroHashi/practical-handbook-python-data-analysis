# pickle形式

## pickleの注意点

```python
import pickle
import pandas as pd
```

```python
df = pd.read_csv(
    "data/FEH_00200524_230205124938.csv",
    encoding="cp932",
    parse_dates=["時間軸（年月日現在）"],
    date_format="%Y年%m月",
    thousands=",",
)
df.head()
```

```python
with open("data/csv-to-pickle.pickle", "wb") as f:
    pickle.dump(df, f)
```

```python
with open("data/csv-to-pickle.pickle", "rb") as f:  # ・・③
    df_loaded = pickle.load(f)  # ・・④
```
