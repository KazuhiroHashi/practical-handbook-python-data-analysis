## テキストファイルとして読み込む

```python
with open("data/FEH_00200524_230205124938.csv", "r", encoding="cp932") as f:  # ・・①
    for i, row in enumerate(f):  # ・・②
        for col in row.strip().split(","):  # ・・③
            print(col, end="|")  # ・・④
        print()
        print("===")
        if i > 1:  # ・・⑤
            break
```

## csvモジュールを使う

```python
import csv

with open(
    "data/FEH_00200524_230205124938.csv",
    "r",
    encoding="cp932",
    newline="",
) as f:
    data = csv.reader(f)
    for i, row in enumerate(data):
        for col in row:
            print(col, end="|")
        print()
        print("===")
        if i > 1:
            break
```
